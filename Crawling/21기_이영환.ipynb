### 과제 url

url => https://movie.naver.com/movie/sdb/rank/rmovie.naver 

평점은 관람객 평점만 하셔도 괜찮습니다!

```python
import requests
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver'
res = requests.get(url)
print(res)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, 'html.parser')
```

```python

# 영화제목 가져오기
title = soup.find_all('div', class_="tit3")

# 영화제목 저장할 리스트
titles = []

# 영화제목 상세페이지 URL 저장
titles_info = []

# 상위 10개 영화제목 추출
for i in range(10):
  titles.append(title[i].find('a').get_text())
  titles_info.append(title[i].find('a')['href'])


# 결과 확인
for text in titles:
  print(text)

for i in range(10):
  titles_info[i] = 'https://movie.naver.com'+titles_info[i] 
  print(titles_info[i])
```

```python
movie_info = list()


for a in range(10):
  movie_res = requests.get(titles_info[a])
  movie_soup = BeautifulSoup(movie_res.text, 'html.parser')
  movie_info_dict = dict()
  movie_info_dict['관람객평점'] = movie_soup.find('span', class_ = 'st_on').get_text()
  movie_info_dict['감독'] = movie_soup.select_one("#content > div.article > div.mv_info_area > div.mv_info > dl > dd:nth-of-type(2) > p")
  movie_info_dict['개요'] = movie_soup.select_one("#content > div.article > div.mv_info_area > div.mv_info > dl > dd:nth-of-type(1) > p")
  movie_info_dict['등급'] = movie_soup.select_one("#content > div.article > div.mv_info_area > div.mv_info > dl > dd:nth-of-type(4) > p > a")
  movie_info_dict['누적관객수'] = movie_soup.select_one("div.mv_info_area div.mv_info dl div p")
  movie_info.append(movie_info_dict)
  
```

```python
movie_info
```

```python
import re

cleanr = re.compile('<.*?>')

for b in range(10):
  movie_info[b]['관람객평점'] =re.sub('관람객 평점 ','',  str(movie_info[b]['관람객평점']))
  movie_info[b]['누적관객수'] = re.sub(cleanr,'', str(movie_info[b]['누적관객수']))
  movie_info[b]['개요'] = re.sub('</span>',',', str(movie_info[b]['개요']))  
  movie_info[b]['개요'] = re.sub(cleanr,'', str(movie_info[b]['개요']))   
  movie_info[b]['감독'] = re.sub(cleanr,'', str(movie_info[b]['감독']))   
  movie_info[b]['등급'] = re.sub(cleanr,'', str(movie_info[b]['등급']))    
  movie_info[b]['개요'] = re.sub('\n|\t|\r','', movie_info[b]['개요'])
```

```python
movie_info
```

```python
import pandas as pd

final_data = pd.DataFrame(index= titles, data = movie_info)
```
