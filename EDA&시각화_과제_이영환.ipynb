### 2) 결측치 확인

- #### <font color='orange'> Q2. 각 컬럼의 결측치를 확인하고, 결측치가 많은 순서대로 결측치 비율과 함께 표로 나타내보세요.
<font color='orange'> (결측치 0개인 컬럼은 나타내지 마세요)

```python
# Q2
# 각 컬럼 결측치 확인
null_num = df.isnull().sum()
# Null 있는 Column만 추출
null_num = null_num[null_num > 0]
# Null 비율 계산
rate = null_num/len(df)*100

# 데이터 프레임 생성
null_num= pd.DataFrame(data = null_num.sort_values(ascending=False),columns = ['Null_num'])
null_rate = pd.DataFrame(data = rate, columns=['missing_%'])

# 표 생성
null_table = pd.concat([null_num,null_rate],axis=1)
null_table
```

- #### <font color='orange'> Q3. 결측치가 있는 컬럼과 결측치 비율을 보기쉽게 그래프로 나타내 보세요.

```python
# Q3
plt.figure(figsize=(15,5))
sns.set_palette("crest")
bar = sns.barplot(data=null_table, x=null_table.index, y='missing_%')
bar.set_xticklabels(bar.get_xticklabels(), rotation=45)
bar
```

- #### <font color='orange'> Q4. 결측치가 적은 beds 컬럼을 좀 더 자세히 살펴보겠습니다. 
  #### <font color='orange'> 'beds' 컬럼이 결측치인 데이터의 상위 5개를 출력해주세요.

```python
# Q4
null_beds = df[df['beds'].isnull()]
null_beds.head()
```

- #### <font color='orange'> Q6. 위 그래프의 세로축이 갯수가 아닌 비율을 나타내도록 재구성해보세요.

```python
# Q6
rate = df['room_type'].value_counts() / df['room_type'].size *100
room_rate = pd.DataFrame(data=rate)

sns.barplot(data=room_rate, x=room_rate.index, y='room_type')
plt.xlabel('room_type')
plt.ylabel('room_rate (%)')
plt.ylim(0,100)
```
